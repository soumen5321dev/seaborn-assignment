```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot
# using x = "timepoint" and y = "signal" for different events and regions.

# Note: timepoint, signal, event, and region are columns in the fmri dataset.

import seaborn as sns

fmri = sns.load_dataset('fmri')
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri)
```

```python
# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots
# using x = 'pclass', y = 'age' and y = 'fare'.

# Note: pclass, age, and fare are columns in the titanic dataset.

titanic = sns.load_dataset('titanic')
sns.boxplot(x = 'pclass', y = 'age', data = titanic)
sns.boxplot(x = 'pclass', y = 'fare', data = titanic)
```

```python
# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column.
# Use the hue parameter for the 'cut' column of the diamonds dataset.

diamonds = sns.load_dataset('diamonds')
sns.histplot(x = 'price', hue = 'cut', data = diamonds)
```

```python
# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the
# "species" column of the iris dataset.

iris = sns.load_dataset('iris')
sns.pairplot(data = iris, hue = 'species')
```

```python
# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

#flights = sns.load_dataset("flights")
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset("flights")

# Pivot the dataset to create a matrix
flights_matrix = flights.pivot("month", "year", "passengers")

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data=flights_matrix, cmap="YlGnBu", annot=True, fmt="d", linewidths=0.5)

# Customize the plot
plt.title('Passenger Counts by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')

# Show the plot
plt.show()
```
